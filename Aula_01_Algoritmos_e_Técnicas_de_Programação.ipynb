```
alunos = ['Ana', 'Bruno', 'Claudio', 'Daniel', 'Fábio', 'José', 'Maria',
          'Nair', 'Paulo', 'Renato'] #criei lista de alunos
notas = [9.0, 4.5, 7.5, 8.0, 10.0, 6.0, 2.5, 8.5, 5.5, 6.5]   #criei lista de notas dos alunos

print("Alunos reprovados")
for nota in range(len(alunos)):
  if notas[nota] < 5:
    print(f"{alunos[nota]:13s} {notas[nota]:2.1f}")

aprovados = []  # criei uma lista vazia
recuperacao = []  # crio outra lista vazia
reprovados = [] #criei lista vazia
for i in range(len(alunos)):
  if notas[i] >= 6:  # aprovados
    aprovados.append(alunos[i])  #  append é a ação de acrescentar um elemento no fim de uma lista, no caso, a "aprovados"
  elif notas[i]  >=5:
    recuperacao.append(alunos[i])  #  append acrescenta um elemento no fim de uma lista "recuperacao"
  else:  # recuperação
    reprovados.append(alunos[i]) #acrescenta um elemento na lista reprovados


print(f'Alunos aprovados: {aprovados}')
print(f'Alunos em recuperação: {recuperacao}')
print(f'Alunos reprovados{reprovados}')
```

```
#Criar um código em Python que calcule o produto escalar de dois vetores, A e B (teste com 2 vetores de tamanho 5):
#produtoEscalar=A[0]∗B[0]+A[1]∗B[1]+A[2]∗B[2]+..+A[N−1]∗B[N−1]

A = [1, 2, 3, 4, 5,10,6]
B = [2, -2, 0, 4, -2, 5,7]


soma = 0
for i in range(len(A)):
  soma = soma + A[i]*B[i]

print(f"O valor do produto escalar é {soma}")
```
